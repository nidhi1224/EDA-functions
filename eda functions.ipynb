```python
# Z-score method for outlier detection and capping
z_scores = np.abs(zscore(df['Mileage']))
threshold = 3
 
# Use .loc to modify the original DataFrame
df.loc[z_scores > threshold, 'Mileage'] = df['Mileage'].quantile(0.99)
 
# IQR method for outlier detection and capping
Q1 = df['Mileage'].quantile(0.25)
Q3 = df['Mileage'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
 
# Use .loc to modify the original DataFrame
df.loc[df['Mileage'] > upper_bound, 'Mileage'] = upper_bound
df.loc[df['Mileage'] < lower_bound, 'Mileage'] = lower_bound
 
print("DataFrame after outlier removal:")
print(df)
```

```python
# Mapping for ordinal encoding
condition_mapping = {
    'Poor': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}
 
# Apply ordinal encoding to Condition column
df['Condition_Encoded'] = df['Condition'].map(condition_mapping)
 
# Apply one-hot encoding to Make column
df = pd.get_dummies(df, columns=['Make'], prefix='Make')
 
print("DataFrame after encoding:")
print(df)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
 
# Assuming X_train and y_train are your training data with NaN values
# Example of using SimpleImputer to impute missing values with mean
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
 
# Initialize LinearRegression model
model = LinearRegression()
 
# Train the model
model.fit(X_train_imputed, y_train)
 
# Predict on test set
X_test_imputed = imputer.transform(X_test)
y_pred = model.predict(X_test_imputed)
 
# Calculate evaluation metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
 
print("R-squared (R2):", r2)
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", rmse)
```
